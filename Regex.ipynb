```python
path= folder / 'COVID-19'
filepath = path / 'COVID.csv'

covid_df= pd.read_csv(filepath)
print(covid_df.head())
```

## Change  the hashtags to lower case 
e.g. replace #Coronavirus with #coronavirus

Create a new column clean_text which has modified hashtags.

```python
import re
covid_df['clean_text']= [(re.sub(r'#[A-Za-z]\w*', lambda x: x.group(0).lower(),
                                 tweetContent)) for tweetContent in 
                         covid_df['Tweet Content']]

print(covid_df['clean_text'].head())
```

## Remove RT from text in column clean_text


```python
# Check the number of 'RT's (Retweets) in Tweet Content
len([tweetContent  for tweetContent in covid_df['clean_text']  if(
    len(re.findall(r'^RT ', str(tweetContent))) > 0)])
```

```python
# Remove the 'RT's from Tweet Content
covid_df['clean_text'] = [re.sub(r'^RT ','', str(tweetContent)) for 
                          tweetContent in covid_df['clean_text']]
```

```python
# Check the number of 'RT's again to confirm their removal
len([tweetContent  for tweetContent in covid_df['clean_text']  if(
    len(re.findall(r'^RT ', str(tweetContent))) > 0)])
```

## Remove URLs and links from text in column clean_text

```python
# Check the number of URLs in the original tweet
len([tweetContent  for tweetContent in covid_df['Tweet Content']  if(
    len(re.findall(r'(https?|ftp):\/\/\S+', str(tweetContent))) > 0)])
```

```python
# Remove the URLs and links (http(s) and ftp links)
covid_df['clean_text'] = [re.sub(r'(https?|ftp):\/\/\S+','',
                                 str(tweetContent)) for tweetContent in
                           covid_df['clean_text']]
```

```python
# Check the number of URLs after removal
len([tweetContent  for tweetContent in covid_df['clean_text']  if(
    len(re.findall(r'(https?|ftp):\/\/\S+', str(tweetContent))) > 0)])
```

## Removing Punctuations from text in column clean_text.


```python
import string
covid_df['clean_text'] = [s.translate(str.maketrans('','', 
                                                    string.punctuation)) for s
                           in covid_df['clean_text']]
```

```python
print(string.punctuation)
print(covid_df['clean_text'].head(10))
```

##  Extract number of Hashtags in a new column num_hashtags


```python
# Twitter hashtags always start with an alphabet and allows alphanumeric 
#characters from second character onwards. 
covid_df['num_hashtags'] = [len(re.findall(r'#[A-Za-z]\w*',
                                           str(tweetContent))) for tweetContent
                             in covid_df['Tweet Content']]
```

```python
print(covid_df.head())
```

## Extract number of user mentions in a new column num_mentions

```python
#twitter allows usernames between 4 and 50 characters. If @ occurs at
# the beginning of a line/sentence or after a space it marks a mention

covid_df['num_mentions'] = [len(re.findall(r'^@([a-zA-Z0-9_]{4,50})', 
                                           tweetContent)) +
                             len(re.findall(r' @([a-zA-Z0-9_]{4,50})', 
                                           tweetContent)) for tweetContent
                             in covid_df['Tweet Content']]
```

```python
print(covid_df[['Tweet Content','num_mentions']][1:10])
```

## Count number of mentions and hashtags

```python
hash_count= covid_df['num_hashtags'].sum()
mention_count= covid_df['num_mentions'].sum()

print('Total number of hashtags= {}'.format(hash_count))
print('Total number of mentions= {}'.format(mention_count))
print('Total number of hashtags and mentions combined = {}'.format(
    hash_count+mention_count))
```
